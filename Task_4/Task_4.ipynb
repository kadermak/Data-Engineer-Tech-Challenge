```python
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt

raw_data = requests.get("https://api.covid19api.com/country/singapore?from=2022-08-01T00:00:00Z&to=2022-08-21T00:00:00Z").json()
df = pd.json_normalize(raw_data, max_level=1)
# df.head()
# df.info()
# plt.legend()

plt.figure(figsize=(12,12))
plt.plot(df['Date'].astype(str), df['Confirmed'], color='g', label='Confirmed (in millions)')
plt.plot(df['Date'].astype(str), df['Deaths'], color='r', label='Deaths')
plt.legend()
plt.show()
```
